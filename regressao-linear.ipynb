## Visualização do algoritmo de Machine Learning Linear Regression com a lib Python para ML e IA scikit-learn.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Carregando os dados e selecionando as váriaveis.

```python
cars = pd.read_csv("cars.csv")

x = cars[['year']].values
y = cars['selling_price'].values
```

Dividindo os dados em conjuntos de treinamento e teste.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

Mostrando os resultados

```python
plt.scatter(x_test, y_test, color='blue', label='Dados Reais')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Regressão Linear')
plt.xlabel('Ano do Carro')
plt.ylabel('Preço de Venda')
plt.title('Regressão Linear - Preço de Venda vs Ano do Carro')
plt.legend()
plt.show()
```
